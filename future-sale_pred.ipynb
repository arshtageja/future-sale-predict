```python
data["date"]=pd.to_datetime(data['date'])
data["year"]=data["date"].dt.year
data["month"]=data["date"].dt.month
data["day"]=data["date"].dt.day
```

```python
data=data.drop(["date_block_num"],axis=1)
```

```python
data=data.sample(frac=0.04)
data.shape
```

```python
target=data["item_cnt_day"]
data=data.drop(["item_cnt_day"], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
```

```python
df=decision_tree.fit(X_train,y_train)
```

```python
res=df.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,res)
```

```python
from sklearn.ensemble import RandomForestClassifier
r=RandomForestClassifier()
```

```python
h=r.fit(X_train,y_train)
res1=h.predict(X_test)
```

```python
accuracy_score(y_test,res1)
```

```python
from sklearn import svm
d=svm.SVC()
r2=d.fit(X_train,y_train)
rest=r2.predict(X_test)
```

```python
accuracy_score(y_test,rest)
```
